```python
import os
from openpyxl import load_workbook

def replace_in_excel_files(root_folder, old_value, new_value):
    for foldername, subfolders, filenames in os.walk(root_folder):
        for filename in filenames:
            if filename.endswith(".xlsx"):
                file_path = os.path.join(foldername, filename)
                try:
                    workbook = load_workbook(file_path)
                    for sheet_name in workbook.sheetnames:
                        sheet = workbook[sheet_name]
                        for row in sheet.iter_rows():
                            for cell in row:
                                if cell.data_type == "s" and old_value in cell.value:
                                    cell.value = cell.value.replace(old_value, new_value)
                    workbook.save(file_path)
                    workbook.close()
                except Exception as e:
                    print(f"Error processing file {file_path}: {str(e)}")

# Example usage:
root_folder = r"C:\Users\avakbal"  # Replace with the path to your main folder
old_value = ""
new_value = ""
replace_in_excel_files(root_folder, old_value, new_value)
```

```python
import os
import xlrd
from xlutils.copy import copy

def replace_in_excel_files(root_folder, old_value, new_value):
    for foldername, subfolders, filenames in os.walk(root_folder):
        for filename in filenames:
            if filename.endswith(".xls"):
                file_path = os.path.join(foldername, filename)
                try:
                    # Read the old .xls workbook
                    workbook = xlrd.open_workbook(file_path, formatting_info=True)
                    # Create a copy to write modifications
                    workbook_copy = copy(workbook)

                    # Iterate through sheets
                    for sheet_index in range(workbook.nsheets):
                        sheet = workbook.sheet_by_index(sheet_index)
                        modified_sheet = workbook_copy.get_sheet(sheet_index)

                        # Iterate through rows and columns
                        for row_index in range(sheet.nrows):
                            for col_index in range(sheet.ncols):
                                cell_value = sheet.cell_value(row_index, col_index)

                                # Check if cell contains old_value
                                if isinstance(cell_value, str) and old_value in cell_value:
                                    # Replace old_value with new_value
                                    new_cell_value = cell_value.replace(old_value, new_value)
                                    # Update the value in the copied workbook
                                    modified_sheet.write(row_index, col_index, new_cell_value)

                    # Save the modified workbook
                    workbook_copy.save(file_path)

                except Exception as e:
                    print(f"Error processing file {file_path}: {str(e)}")

# Example usage:
root_folder = r"C:\Users\avakbal"  # Replace with the path to your main folder
old_value = ""
new_value = ""
replace_in_excel_files(root_folder, old_value, new_value)
```
